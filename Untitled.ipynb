## Data preprocessing

#### Use regular expression to clean our text

```python
import re
```

```python
def textCleaning(text):
   
    text = text.lower() #lower case
    text = re.sub('\[.*?\]','',text) # remove text in squre backets
    text = re.sub('https?://\S+|www\.\S+','',text) # remove URL 
    text = re.sub('[%s]' % re.escape(string.punctuation),'',text)# remove punctuation and special characters
    text = re.sub('\n','',text) # remove line breaks
    text = re.sub('\w*\d\w*','',text) # remove words containing numbers
    emoji_pattern = re.compile("["
                           u"\U0001F600-\U0001F64F"  # emoticons
                           u"\U0001F300-\U0001F5FF"  # symbols & pictographs
                           u"\U0001F680-\U0001F6FF"  # transport & map symbols
                           u"\U0001F1E0-\U0001F1FF"  # flags (iOS)
                           u"\U00002702-\U000027B0"
                           u"\U000024C2-\U0001F251"
                           "]+", flags=re.UNICODE)
    text=emoji_pattern.sub(r'', text) #remove Emojis
    return text
```

```python
train['text'] = train['text'].apply(textCleaning)
```

```python
test['text'] = test['text'].apply(textCleaning)
```

```python
from nltk.tokenize import sent_tokenize,word_tokenize
from nltk.stem.wordnet import WordNetLemmatizer
lemmatizer = WordNetLemmatizer()

for i in range(len(tweets_train)):
    
    sentences = sent_tokenize(tweets_train[i])
    
    word_list = []
    for sent in sentences:
        
        words = word_tokenize(sent)
        
        for word in words:
            if words not in word_list:
                word_list.append(word)
                
    
    word_list = [lemmatizer.lemmatize(w,'v') for w in word_list if w not in stopWords]
    
    tweets_train[i] = ' '.join(w for w in word_list)
    
for i in range(len(tweets_test)):
    
    sentences = sent_tokenize(tweets_test[i])
    
    word_list = []
    for sent in sentences:
        
        words = word_tokenize(sent)
        
        for word in words:
            
            if words not in word_list:
                word_list.append(word)
                
    
    word_list = [lemmatizer.lemmatize(w,'v') for w in word_list if w not in stopWords]
    
    tweets_test[i] = ' '.join(w for w in word_list)
```
